```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
  return 2*x
```

```python
a=0
b=1
M=2
```

```python
def aceptacion (n):
  lista=[]
  for _ in range(n):
    R1=rd.random()
    R2=rd.random()
    x=a+((b-a)*R1)
    if R2<=f(x)/M:
      lista.append(x)
  return lista
```

```python
c=aceptacion(10000)
```

```python
print(c)
```

```python
print(np.var(c))
```

```python
print(np.mean(c))
```

```python
plt.hist(c, bins=50)
plt.show()
```
