## Distribución Normal


**Nomenclatura estándar:**  
X ~ N(μ, σ²)

**Función de densidad:**  
f(x) = (1 / (σ√(2π))) * exp(- (x - μ)² / (2σ²))

**Función de distribución acumulativa:**  
F(x) = ∫ f(t) dt desde -∞ hasta x

**Media:** μ  
**Varianza:** σ²  
**Desviación estándar:** σ

## Distribución Uniforme (Continua)


**Nomenclatura estándar:**  
X ~ U(a, b)

**Función de densidad:**  
f(x) = 1 / (b - a) para a ≤ x ≤ b

**Función de distribución acumulativa:**  
F(x) = (x - a) / (b - a) para a ≤ x ≤ b

**Media:** (a + b) / 2  
**Varianza:** (b - a)² / 12  
**Desviación estándar:** √Varianza


```python

from scipy.stats import uniform

a, b = 0, 10
x = np.linspace(a - 2, b + 2, 1000)
pdf = uniform.pdf(x, a, b - a)
cdf = uniform.cdf(x, a, b - a)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x, pdf)
plt.title("Uniforme Continua - Densidad")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, cdf, color='orange')
plt.title("Uniforme Continua - Acumulativa")
plt.grid(True)
plt.show()

samples = np.random.uniform(a, b, 1000)
```

**Seudocódigo para la simulación:**


Entrada: límites a, b y n muestras  
Para i de 1 a n:  
 Generar u uniformemente en (0, 1)  
 Calcular x = a + (b - a) * u  
 Guardar x  
Retornar muestras


## Distribución Triangular


**Nomenclatura estándar:**  
X ~ Triangular(a, c, b)

**Función de densidad:**  
f(x) =
- 2(x - a) / ((b - a)(c - a)) para a ≤ x ≤ c  
- 2(b - x) / ((b - a)(b - c)) para c < x ≤ b

**Media:** (a + b + c) / 3  
**Varianza:** (a² + b² + c² - ab - ac - bc) / 18  
**Desviación estándar:** √Varianza


```python

from scipy.stats import triang

a, b, c = 0, 10, 6
c_scaled = (c - a) / (b - a)
x = np.linspace(a, b, 1000)
pdf = triang.pdf(x, c_scaled, loc=a, scale=b - a)
cdf = triang.cdf(x, c_scaled, loc=a, scale=b - a)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x, pdf)
plt.title("Distribución Triangular - Densidad")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, cdf, color='orange')
plt.title("Distribución Triangular - Acumulativa")
plt.grid(True)
plt.show()

samples = triang.rvs(c_scaled, loc=a, scale=b - a, size=1000)
```

**Seudocódigo para la simulación:**


Entrada: parámetros a, c, b y n muestras  
Para i de 1 a n:  
 Generar u en (0, 1)  
 Si u < (c - a) / (b - a):  
  x = a + sqrt(u * (b - a)(c - a))  
 Sino:  
  x = b - sqrt((1 - u)(b - a)(b - c))  
 Guardar x  
Retornar muestras


## Distribución Exponencial


**Nomenclatura estándar:**  
X ~ Exp(λ)

**Función de densidad:**  
f(x) = λ * e^(-λx) para x ≥ 0

**Función de distribución acumulativa:**  
F(x) = 1 - e^(-λx)

**Media:** 1 / λ  
**Varianza:** 1 / λ²  
**Desviación estándar:** 1 / λ


```python

from scipy.stats import expon

lam = 1
x = np.linspace(0, 10, 1000)
pdf = expon.pdf(x, scale=1/lam)
cdf = expon.cdf(x, scale=1/lam)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x, pdf)
plt.title("Distribución Exponencial - Densidad")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, cdf, color='orange')
plt.title("Distribución Exponencial - Acumulativa")
plt.grid(True)
plt.show()

samples = np.random.exponential(1/lam, 1000)
```

**Seudocódigo para la simulación:**


Entrada: tasa λ y n muestras  
Para i de 1 a n:  
 Generar u uniformemente en (0, 1)  
 Calcular x = -ln(1 - u) / λ  
 Guardar x  
Retornar muestras


## Distribución m-Erlang


**Nomenclatura estándar:**  
X ~ Erlang(k, λ)

**Función de densidad:**  
f(x) = (λ^k * x^(k-1) * e^(-λx)) / (k - 1)!

**Función de distribución acumulativa:**  
F(x) = 1 - Σ_{n=0}^{k-1} ((λx)^n / n!) * e^(-λx)

**Media:** k / λ  
**Varianza:** k / λ²  
**Desviación estándar:** √(k) / λ


```python

from scipy.stats import erlang

k, lam = 3, 1
x = np.linspace(0, 15, 1000)
pdf = erlang.pdf(x, k, scale=1/lam)
cdf = erlang.cdf(x, k, scale=1/lam)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x, pdf)
plt.title("Distribución m-Erlang - Densidad")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, cdf, color='orange')
plt.title("Distribución m-Erlang - Acumulativa")
plt.grid(True)
plt.show()

samples = erlang.rvs(k, scale=1/lam, size=1000)
```

**Seudocódigo para la simulación:**


Entrada: parámetros k, λ y n muestras  
Para i de 1 a n:  
 Inicializar x = 0  
 Para j de 1 a k:  
  Generar u uniformemente en (0, 1)  
  x += -ln(1 - u) / λ  
 Guardar x  
Retornar muestras

## Distribución Weibull


**Nomenclatura estándar:**  
X ~ Weibull(k, λ)

**Función de densidad:**  
f(x) = (k / λ) * (x / λ)^(k - 1) * e^(-(x / λ)^k)

**Media:** λ * Γ(1 + 1/k)  
**Varianza:** λ² * (Γ(1 + 2/k) - (Γ(1 + 1/k))²)  
**Desviación estándar:** √Varianza


```python

from scipy.stats import weibull_min

k, lam = 1.5, 1
x = np.linspace(0, 5, 1000)
pdf = weibull_min.pdf(x, k, scale=lam)
cdf = weibull_min.cdf(x, k, scale=lam)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x, pdf)
plt.title("Distribución Weibull - Densidad")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, cdf, color='orange')
plt.title("Distribución Weibull - Acumulativa")
plt.grid(True)
plt.show()

samples = weibull_min.rvs(k, scale=lam, size=1000)
```

**Seudocódigo para la simulación:**


Entrada: parámetros k, λ y n muestras  
Para i de 1 a n:

##Distribución Jonhson

**Nomenclatura estándar:**  
X ~ Johnson SU(γ, δ, ξ, λ)

**Transformación:**  
Z = γ + δ * sinh⁻¹((X - ξ)/λ), donde Z ~ N(0, 1)

**Media, varianza, desviación estándar:** calculadas numéricamente

```python
from scipy.stats import johnsonsu

gamma_param, delta_param, xi, lambd = 1.5, 1.2, 0, 1
x = np.linspace(-5, 5, 1000)
pdf = johnsonsu.pdf(x, gamma_param, delta_param, loc=xi, scale=lambd)
cdf = johnsonsu.cdf(x, gamma_param, delta_param, loc=xi, scale=lambd)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x, pdf)
plt.title("Distribución Johnson SU - Densidad")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, cdf, color='orange')
plt.title("Distribución Johnson SU - Acumulativa")
plt.grid(True)
plt.show()

samples = johnsonsu.rvs(gamma_param, delta_param, loc=xi, scale=lambd, size=1000)
```

**Seudocódigo:**

## Distribución Uniforme (Discreta)


**Nomenclatura estándar:**  
X ~ U({a, ..., b})

**Función de masa de probabilidad:**  
P(X = k) = 1 / (b - a + 1)

**Media:** (a + b) / 2  
**Varianza:** ((b - a + 1)² - 1) / 12  
**Desviación estándar:** √Varianza


```python
a, b = 1, 6
x = np.arange(a, b + 1)
pmf = [1/(b - a + 1)] * len(x)
cdf = np.cumsum(pmf)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
# Removed the deprecated 'use_line_collection' argument
plt.stem(x, pmf)
plt.title("Uniforme Discreta - PMF")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.step(x, cdf, where='post', color='orange')
plt.title("Uniforme Discreta - CDF")
plt.grid(True)
plt.show()

samples = np.random.randint(a, b+1, 1000)
```

## Distribución de Poisson


**Nomenclatura estándar:**  
X ~ Poisson(λ)

**Función de masa de probabilidad:**  
P(X = k) = (λ^k * e^(-λ)) / k!

**Función de distribución acumulativa:**  
F(k) = ∑ P(X = i), para i = 0 hasta k

**Media:** λ  
**Varianza:** λ  
**Desviación estándar:** √λ


```python
from scipy.stats import poisson

lam = 3
x = np.arange(0, 15)
pmf = poisson.pmf(x, lam)
cdf = poisson.cdf(x, lam)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
# Removed the deprecated 'use_line_collection' argument
plt.stem(x, pmf)
plt.title("Distribución de Poisson - PMF")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.step(x, cdf, where='post', color='orange')
plt.title("Distribución de Poisson - CDF")
plt.grid(True)
plt.show()

# Simulación
samples = np.random.poisson(lam, 1000)
```

**Seudocódigo para la simulación:**


Entrada: tasa λ, número de muestras n  
Para i de 1 a n:  
 Inicializar L = exp(-λ), p = 1, k = 0  
 Mientras p > L:  
  k = k + 1  
  Generar u uniformemente en (0, 1)  
  p = p * u  
 Guardar k - 1  
Retornar muestras

##Distribución Binomial Negativa

**Nomenclatura estándar:**
X ~ NB(r, p)

**Función de masa:**  
P(X = k) = C(k + r - 1, k) * (1 - p)^k * p^r, k ∈ {0, 1, 2, ...}

**Media:*** (r * (1 - p)) / p  
**Varianza:** (r * (1 - p)) / p²  
**Desviación estándar:** √Varianza

```python
from scipy.stats import nbinom
import matplotlib.pyplot as plt
import numpy as np
r, p = 5, 0.4
x = np.arange(0, 30)
pmf = nbinom.pmf(x, r, p)
cdf = nbinom.cdf(x, r, p)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.stem(x, pmf, basefmt=" ")
plt.title("Binomial Negativa - Función de Masa")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.step(x, cdf, where='post', color='orange')
plt.title("Binomial Negativa - Función Acumulativa")
plt.grid(True)
```
