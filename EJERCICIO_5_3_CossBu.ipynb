### Paso 4:

### Paso 5
Se diseña el experimento, generando numeros aleatorios

```python
#PASO 5
def experimento(n=10):
  exitos=0
  for i in range(10):
    X1=np.random.normal(1.5,0.04,1)
    X2=np.random.normal(1.48,0.03,1)
    if X2>=X1:
      exitos+=1
  prob=exitos/n
  return prob
```

### Paso 6:
Repetir m veces el experimento

```python
#PASO 6
def muestra(m=1000,n=10):
  lista=[]
  for i in range(m):
    e=experimento(n=n)
    lista.append(e)
  return lista
```

### Paso 7:
Se visualizan los resultados para determinar el tamaño de la muestra necesaria

```python
#PASO 7
M=muestra(m=1000)
plt.plot(M)
plt.show()
```

### Paso 8:
Función simulación

```python
#PASO 8
#DEVONVER LA FUNCIÓN SIMULACIÓN DONDE DE UN VALOR DE PROBABILIDAD
#es decir ahorita es una corrida de m con 1000 y n con 10
#buscamos que haya mas corridas para hacer estadística con histograma
def simulación(corridas=500, m=1000, n=10):
  resultados=[]
  for _ in range(corridas):
    estimaciones = muestra(m=m, n=n)
    media_estimaciones = np.mean(estimaciones)
    resultados.append(media_estimaciones)

  plt.hist(resultados, bins=30)
  plt.show()
  return resultados

resultados=simulación(corridas=500, m=1000, n=10)
```
